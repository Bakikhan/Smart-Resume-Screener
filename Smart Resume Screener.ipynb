```python
def Clean(df):
    df=re.sub(r'@[A-Za-z0-9]+','',df) # Remove @ functions
    df=re.sub(r"(?:\@|https?\://)\S+", " ", df)
    df=re.sub(r'[^\x00-\x7f]',r'', df) # Remove Ascii 
    df=re.sub(r",+"," ",df)
    list1="-():!*?\\//\r\n.--[]\'"
    for alt in list1:
        df = df.replace(alt," ").lower()
    df = df.replace("   "," ").replace("  "," ")
    #df=df.replace('\r', ' ').replace('\n', ' ').replace('!', ' ').replace('?', ' ').replace("  "," ").lower() #remove \n, \r, '!' and lowercase
    return df
```
